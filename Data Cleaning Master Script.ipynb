```python
import pandas as pd
import numpy as np
import os
import time
import random

def DataCleaningMaster(datapath, dataname):
    print(f"Thank you for giving the details!")

    sec = random.randint(1, 4)
    print(f"Please wait for {sec} seconds! Checking file path...")
    time.sleep(sec)

    if not os.path.exists(datapath):
        print("Please enter correct path!")
        return

    # Checking file type
    if datapath.endswith('.csv'):
        print('Dataset is CSV!')
        data = pd.read_csv(datapath)
    elif datapath.endswith('.xlsx'):
        print('Dataset is Excel file!')
        data = pd.read_excel(datapath)
    else:
        print('Unknown file type!')
        return

    print(f"Please wait for {sec} seconds! Checking total columns and rows...")
    time.sleep(sec)
    print(f"Dataset contains total rows: {data.shape[0]} and columns: {data.shape[1]}")

    print(f"Please wait for {sec} seconds! Checking duplicates...")
    time.sleep(sec)
    duplicates = data.duplicated()
    totalduplicate = duplicates.sum()
    print(f"Dataset has {totalduplicate} duplicate records.")

    if totalduplicate > 0:
        print(f"Saving {totalduplicate} duplicate records...")
        duplicaterecords = data[duplicates]
        duplicaterecords.to_csv(f"{dataname}_duplicates.csv", index=None)

    # Remove duplicates
    df = data.drop_duplicates()

    print(f"Please wait for {sec} seconds! Checking missing values...")
    time.sleep(sec)
    totalmissingvalues = df.isnull().sum().sum()
    missingvaluebycolumns = df.isnull().sum()
    print(f"Dataset has total missing values: {totalmissingvalues}")
    print("Missing values by column:\n", missingvaluebycolumns)

    print(f"Please wait for {sec} seconds! Cleaning dataset...")
    time.sleep(sec)

    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            df[col] = round(df[col].fillna(df[col].mean()), 2)
        else:
            df.dropna(subset=[col], inplace=True)

    print(f"Please wait for {sec} seconds! Exporting dataset...")
    time.sleep(sec)
    print(f"Congrats! Dataset cleaned — Rows: {df.shape[0]}, Columns: {df.shape[1]}")

    df.to_csv(f"{dataname}_Clean_data.csv", index=None)
    print("Clean dataset saved successfully!")

    return df
```
